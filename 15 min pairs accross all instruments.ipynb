```python
import pandas as pd
import numpy as np
import statsmodels
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller

import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style="whitegrid")
import matplotlib.pylab as plt
import os
```

```python
instruments=os.listdir("15m")
instruments
```

```python
first=instruments[0]
second=instruments[1]
first_df=pd.read_csv("15m/"+first)
second_df=pd.read_csv("15m/"+second)
strategy=pd.merge(first_df,second_df,on='time',suffixes=('_1', '_2'))[['time','close_1','close_2']].dropna(how='any')
_,p,_=coint(strategy['close_1'],strategy['close_2'])
print(p,first,second)
```

```python
def print_coint(first,second):
    first_df=pd.read_csv("15m/"+first)
    second_df=pd.read_csv("15m/"+second)
    strategy=pd.merge(first_df,second_df,on='time',suffixes=('_1', '_2'))[['time','close_1','close_2']].dropna(how='any')
    _,p,_=coint(strategy['close_1'],strategy['close_2'])
    print(p,first,second)
```

```python
print_coint(instruments[0],instruments[1])
```
